```python
#Calculate the homework scores by summing the actual scores and dividing by the total possible points.
homework_scores = [col for col in final_df.columns if col.startswith('Homework') and 'Max Points' not in col]
homework_max_points = [col for col in final_df.columns if col.startswith('Homework') and 'Max Points' in col]

final_df['Total Homework Score (%)'] = (final_df[homework_scores].sum(axis=1) / final_df[homework_max_points].sum(axis=1)) * 100

final_df[['Total Homework Score (%)'] + homework_scores].head()
```

```python
#Calculate the average homework scores by dividing each actual score by its corresponding maximum points, summing these ratios, and then dividing by the number of assignments.
average_homework_score = (
    (final_df[homework_scores].values / final_df[homework_max_points].values).sum(axis=1)
    / len(homework_scores)
) * 100
```

## Group the data to calculate final scores!

```python
import numpy as np

expected_columns = ['Exam 1 Score', 'Exam 2 Score', 'Exam 3 Score', 'Quiz Score', 'Homework Score']
actual_columns = final_df.columns

if 'Final Quiz Score (%)' in actual_columns:
    final_df = final_df.rename(columns={'Final Quiz Score (%)': 'Quiz Score'})
if 'Final Homework Score (%)' in actual_columns:
    final_df = final_df.rename(columns={'Final Homework Score (%)': 'Homework Score'})

missing_columns = [col for col in expected_columns if col not in final_df.columns]
if missing_columns:
    print(f"Missing columns in final_df: {missing_columns}")
else:
    weights = pd.Series(
        {
            "Exam 1 Score": 0.05,
            "Exam 2 Score": 0.10,
            "Exam 3 Score": 0.15,
            "Quiz Score": 0.30,
            "Homework Score": 0.40,
        }
    )

    final_df['Final Score'] = (final_df[list(weights.index)] * weights).sum(axis=1)

    final_df['Ceiling Score'] = np.ceil(final_df['Final Score'] * 100)

    def convert_to_letter_grade(score):
        if score >= 90:
            return 'A'
        elif score >= 80:
            return 'B'
        elif score >= 70:
            return 'C'
        elif score >= 60:
            return 'D'
        else:
            return 'F'

    final_df['Letter Grade'] = final_df['Ceiling Score'].apply(convert_to_letter_grade)

    final_df['Letter Grade'] = pd.Categorical(final_df['Letter Grade'], categories=['A', 'B', 'C', 'D', 'F'], ordered=True)

    print(final_df[['Ceiling Score', 'Letter Grade']].head())
```

2. Calculate the final score by applying the weights to each component and summing the results.
3. Calculate the ceiling score by rounding up the final score multiplied by 100.
4. Define a function to convert numeric scores to letter grades based on standard thresholds.
5. Map the ceiling scores to letter grades.
6. Store the letter grades as a categorical column in the DataFrame.
